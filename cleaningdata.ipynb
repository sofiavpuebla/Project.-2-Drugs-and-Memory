```python
import pandas as pd
```

```python
data=pd.read_csv("input/data-final.csv", sep="\t" )
```

```python
data.columns
```

```python
general=data[['EXT1', 'EXT2', 'EXT3', 'EXT4', 'EXT5', 'EXT6', 'EXT7', 'EXT8', 'EXT9',
       'EXT10', 'AGR1', 'AGR2', 'AGR3', 'AGR4', 'AGR5', 'AGR6', 'AGR7', 'AGR8', 'AGR9',
       'AGR10','EST1', 'EST2', 'EST3', 'EST4', 'EST5', 'EST6', 'EST7', 'EST8', 'EST9',
       'EST10', 'OPN1', 'OPN2', 'OPN3', 'OPN4', 'OPN5', 'OPN6', 'OPN7', 'OPN8', 'OPN9',
       'OPN10','CSN1', 'CSN2', 'CSN3', 'CSN4', 'CSN5', 'CSN6', 'CSN7', 'CSN8', 'CSN9',
       'CSN10', 'country']]

general
```

```python
general=general[(general["country"].isnull()==0) & (general["country"]!="NONE")]
```

Como vamos a enriquecer el dataframe con los datos por países, aquellas filas que not tienen un país asignado las vamos a eliminar

```python
country=pd.read_csv("input/code.csv")
```

```python
country=country.rename(columns={"Code":"country"})
```

```python
analisis_paises=pd.merge(general, country, how="left",left_on="country", right_on="country")
```

```python
analisis_paises=analisis_paises.rename(columns={"country":"Code", "Name":"Country"})
analisis_paises.head()
```

```python
analisis_paises.to_csv("output/atrib_pais.csv")
```
